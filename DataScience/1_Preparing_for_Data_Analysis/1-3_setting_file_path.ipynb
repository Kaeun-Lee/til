## %mv or %move
- %move : 파일 경로 옮기기
- 다운받은 파일을 현재 작업하고 있는 폴더로 가져오기
    - %mv /Users/~~~/~~~/도로교통공단_교통사고다발지역_20191010.csv .
    - 파일을 다운 받은 폴더의 경로를 적기
    - 파일을 다운 받은 위치의 경우 개인마다 다르기 때문에 자신에 맞게 작성하기
    - 제일 끝에 한 칸을 띄우고 .(마침표)을 꼭 적어야 함

```python
# C드라이브에 file-path 폴더로 이동됨
# C드라이브가 아닌 다른곳(목적지)로 이동하고 싶으면 원하는 경로를 따옴표로 묶어 추가
# 즉, 윈도우 환경일 때 !move "파일 다운로드 한 경로" "보내고 싶은 목적지 경러" .(마침표)한다. (큰따옴표 필수)

!move "C:\Users\Downloads\서울특별시_송파구_종량제봉투가격_20210219_1614734444298_1573.csv" .  # . : 파일을 같은 경로에 있는 폴더로 옮김
```

## pd.read_csv, cp949
- pd.read_csv : pandas로 csv 파일을 불러옴
- UnicodeDecodeError
    - 이 file의 encoding이 utf-8이 아니라는 의미
    - 엑셀에서 저장했다면 한글의 encoding은 cp949로 되어있음
- csv 파일을 불러올 때 한글 파일 인코딩을 하지 않으면 오류가 나기 때문에, encoding="cp949"를 함께 적어야 함
- "euc-kr"을 적는 방법도 있지만 일부 한글은 깨지기 때문에(2350자만 표현, 일부는 깨져서 출력) "cp949"를 사용하는 것을 추천

```python
import pandas as pd

pd.read_csv("./data/서울특별시_송파구_종량제봉투가격_20210219_1614734444298_1573.csv", encoding="cp949")
```
