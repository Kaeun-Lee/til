```python
# 2. select와 for문으로 src값을 모두 가져온 뒤 저장하기
import dload # URL로부터 파일 다운로드를 해주는 패키지
from bs4 import BeautifulSoup
from selenium import webdriver
import time

driver = webdriver.Chrome('chromedriver') # 웹드라이버 파일의 경로
driver.get("https://search.daum.net/search?w=img&nil_search=btn&DA=NTB&enc=utf8&q=%EC%95%84%EC%9D%B4%EC%9C%A0")
time.sleep(5) # 5초 동안 페이지 로딩 기다리기

req = driver.page_source
# HTML을 BeautifulSoup이라는 라이브러리를 활용해 검색하기 용이한 상태로 만듦
# soup이라는 변수에 "파싱 용이해진 html"이 담긴 상태가 됨
# 이제 코딩을 통해 필요한 부분을 추출하면 된다.
soup = BeautifulSoup(req, 'html.parser')

i = 1
thumbnails = soup.select('#imgList > div > a > img')  # 고유값을 지워주고 전체를 가져오도록 한다
# select로 하면 이게 리스트로 딱 떨어지기 때문에 바로 for문을 돌려서 하나 하나에 대한 태그를 볼 수 있다.
for thumbnail in thumbnails:
    img = thumbnail['src']
    dload.save(img, f'img/{i}.jpg') # img 폴더에 'img/1.jpg'라는 형식으로 저장해라
    i += 1

driver.quit() # 끝나면 닫아주기
```

```python
# 가져온 것을 excel 파일로 저장하기
rom openpyxl import Workbook

from bs4 import BeautifulSoup
from selenium import webdriver


driver = webdriver.Chrome('chromedriver')

url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=추석"

driver.get(url)
req = driver.page_source
soup = BeautifulSoup(req, 'html.parser')

articles = soup.select('#main_pack > div.news.mynews.section._prs_nws > ul > li')

wb = Workbook()  # Workbook을 새로 하나 만들어서
ws1 = wb.active  # sheet 제목을 'articles'로 하고
ws1.title = "articles"
ws1.append(["제목", "링크", "신문사"])  # 첫줄에 들어감

for article in articles:
    title = article.select_one('dl > dt > a').text  # article 안에서 또 찾은 후 text만 가져옴
    url = article.select_one('dl > dt > a')['href']  # 기사 url 가져오기
    comp = article.select_one('span._sp_each_source').text.split(' ')[0].replace('언론사','') # 언론사만 가져오기
    # span에서 가장 끝에 걸 가져올지, 좀 더 앞에 걸 가져올지는 실행해보면서 결정하기!

    ws1.append([title, url, comp])  # 둘째줄부터 들어감

driver.quit()
wb.save(filename='articles.xlsx')    # 파일 저장
```

```python
# 첨부 파일 & 여러사람에게 보내기(반복문)
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email import encoders


# 보내는 사람 정보
me = "***@****"
my_password = "*****@***"

# 로그인하기
s = smtplib.SMTP_SSL('smtp.gmail.com')
s.login(me, my_password)

# 받는 사람 정보
emails = ['*****@****', '****@*****'] # 여러 사람 email 담기

for you in emails:
    # 메일 기본 정보 설정
    msg = MIMEMultipart('alternative')
    msg['Subject'] = "[공유] 추석기사"
    msg['From'] = me
    msg['To'] = you

    # 메일 내용 쓰기
    content = "과연 잘 보내졌을까?!"
    part2 = MIMEText(content, 'plain')
    msg.attach(part2)

    # 파일을 읽어서 첨부하기
    part = MIMEBase('application', "octet-stream")
    with open("articles.xlsx", 'rb') as file:
        part.set_payload(file.read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', "attachment", filename="추석기사.xlsx")
    msg.attach(part)

    # 메일 보내고 서버 끄기
    s.sendmail(me, you, msg.as_string())
s.quit()
```
