### 여러 그래프 그리기(하나의 그래프 창에 여러 그래프 그리기)

```````````````python
# plt.plot(x1,y1[,ft1],x2,y2[,fmt2],x3,y3[,fmt3],...,xn,yn[,fmtn])
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5,5,0.1)
y1=2*x**2
y2=5*x+30
y3=20*np.cos(x)**2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
# plt.plot(x,y1,x,y2,x,y3)랑 동일한 결과

plt.show()
```````````````

### 새로운 그래프 창에 그래프 그리기

```````````````python
# 여러 항복의 데이터를 서로 비교할 때 사용
#plt.bar(x, height[,width=width_f,color=colors, tick_label=tick_labels, align='center'(기본) 혹은 'edge'])

# x는 x축에 표시될 위치, heigh는 막대 그래프 데이터

import matplotlib.pyplot as plt
import numpy as np

member_IDs =['m_01','m_02','m_03','m_04']   # 회원 ID
before_ex =[30,40,42,35]        # 운동 시작 전
after_ex=[27,35,40,33]          # 운동 한 달 후

n_data = len(member_IDs)   # 회원이 네 명이므로 전체 데이터 수는 4
index = np.arange(n_data)  # numpy를 이용해 배열 생성 (0,1,2,3)
plt.bar(index, before_ex)  # x=index, height=before_ex로 지정
plt.show()       # x축의 index 위치에서 가운데 막대그래프 표시
```````````````

```````````````python
colors =['r','g','b','m']
plt.bar(index, before_ex, width =0.6, color=colors, tick_label=member_IDs)
# x축 표시 위치에 tick_label로 각각 막대그래프 이름 지정
plt.show()
```````````````

### 곡명 데이터 가져오기

```````````````python
## 빌보드차트로부터 곡명 태그요소 가져오기

import requests
from bs4 import BeautifulSoup

url ="https://www.billboard.com/charts/hot-100"

html_music = requests.get(url).text  # 요청 사이트 html소스 가져오기
soup_music = BeautifulSoup(html_music,'lxml') # html 데이터 파싱하여 soup_music에 저장

# 파싱된 데이터 soup_music으로부터 곡명 데이터가 위치한 태그 요소 추출
titles= soup_music.select('span.chart-element__information__song')
titles
```````````````

```````````````python
## 태그 요소 항목으로부터 곡명(텍스트) 추출

music_titles =[]    # 추출한 곡명을 담을 리스트

for title in titles:
    st_title = title.get_text().strip() # 문자열(텍스트)만 가져온 뒤 앞 뒤 개행문자나 공백 제거
    music_titles.append(st_title)   # music_titles 리스트에 한 곡명씩 추가
    
music_titles
```````````````

### 아티스트 데이터 가져오기

```````````````python
##파싱된 데이터에서 아티스트 태그요소 가져오고 아티스트(텍스트)추출하기

# 앞에서 파싱된 데이터 soup_music으로부터 아티스트 데이터가 위치한 태그 요소 추출

artists= soup_music.select('span.chart-element__information__artist')
artists

music_artists =[]
for artist in artists:
    st_artist= artist.get_text().strip()  # 텍스트만 가져온 뒤 앞 뒤 개행문자나 공백 제거
    music_artists.append(st_artist)  
    
music_artists
```````````````

## 웹 API 란?

```````````````python
# 웹 API(Wep Application Programming Interface)
- 페이스북, 트위터, 유튜브 등의 SNS와 네이버, 카카오 등의 포털사이트는
  자사가 제공하는 서비스와 다른 회사나 다양한 기기들과 데이터를 공유하기 위해
  API를 제공
- 웹 API는 어떤 기기가 웹 서버에 데이터를 요청하면 웹서버가 다양한 형식
  (JSON, XML 등)으로 응답하는 방식으로 데이터를 주고 받음
- 많은 웹 사이트들이 웹 API를 공개하고 있어, 웹 API를 통해 원하는 데이터를
  요청하여 가져올 수 있음
    
- 웹 사이트에서 어떤 API를 제공하는지를 정리한 사이트
 https://www.programmableweb.com/category/all/apis
```````````````

## 웹 사이트 주소 생성 시 주의할 점

### URL 인코딩(URL encoding)

```````````````python
- 요청주소와 요청 매개변수(params)를 분리해 'requests.get()'에 입력 시 주의

- URL 인코딩은 퍼센트 인코딩(Percent-encoding)이라고도 함
- URL에 사용하는 문자를 인코딩(다른 값으로 변환)하는 방법
- URL 인코딩에서는 알파벳이나 숫자, 몇몇 기호를 제외한 나머지는 16진수 값으로
  변환(인코딩)해 이용함
(예) '+'는 '%2B'로, '/'는 '%2F'로, '='는 '%3D')

- requests.get(url, params=req_parameter)는 URL 인코딩이 자동으로 수행됨
 -- 즉 req_parameter로 들어가는 문자열은 인코딩되지 않은 문자열로 넣어야 함
 -- 그렇지 않을 경우 잘못 생성된 URL로 요청을 보냄
```````````````

## XML 구조

### XML 트리

```````````````python
- XML 문서는 HTML 문서와 마찬가지로 트리(tree) 형태의 계층 구조를 가짐
- XML 트리는 하나뿐인 루트(root) 요소부터 시작하며, 각각의 자식(child)요소에
  차례대로 연결됨
- XML 트리 구조에 포함되는 모든 요소는 자신만의 자식(child) 요소를 가질 수 있음

- 부모(parent)요소는 여러 개의 자식(child) 요소를 가질 수 있음
  하지만 자식(child) 요소는 단 하나의 부모(parent) 요소만을 가짐
- 형제(sibling) 요소는 같은 트리 레벨(tree level)에 존재하는 요소를 가리킴
  즉, 형제 요소란 같은 부모(parent) 요소를 가지는 자식 요소 간의 관계를 나타냄
    
- 조상(ancestor) 요소란 부모 요소를 포함해 계층적으로 현재 요소보다 상위에 있는 모든 요소를 가리킴
- 자손(descendant) 요소란 자식 요소를 포함해 계층적으로 현재 요소보다 하위에 있는 모든 요소를 가리킴

- XML 트리의 모든 요소는 자신만의 텍스트나 속성을 가질 수 있음
```````````````

### XML 선언

```````````````python
- XML 문서는 맨 첫 줄에 <xml>태그를 사용하여 XML 문서임을 명시해야 함
- 이것을 XML 프롤로그(prolog)라고 하며, 이때 사용되는 <xml>태그의 이름은
  소문자 xml로만 사용해야 함
    
- XML 프롤로그의 문법
<?xml version = "XML문서버전" encoding="문자셋" standalone="yes|no"?>

- version 속성 : XML 문서에 사용된 XML의 버전을 명시
  encoding 속성 : XML문서의 문자셋(character set)을 명시하며, 기본값은 UTF-8로 설정됨
  standalone 속성 : XML 문서가 외부 DTD(Document Type Definition)와 같은 외부 소스의 데이터에
                   의존하고 있는 문서인지 아닌지를 XML 파서(parser)에 알려주는 역할을 함
                   기본값은 no이며, yes로 설정하면 이 문서를 파싱(parsing)할 때 참조해야 할
                   외부 소스가 없다는 것을 의미
```````````````

### XML 문법

```````````````python
1. 모든 XML 요소는 종료 태그를 가져야 함
  HTML은 종료 태그를 생략하거나, 빈 태그를 사용해도 대부분 문제없이 동작하지만
  XML에서는 종료 태그가 생략되면 오류가 발생함
  또한, 빈 태그에도 반드시 슬래시(/)를 추가해야만 오류가 발생하지 않음

    예제
HTML : <h1>XML
       <hr>

XML  : <h1>XML</h1>
       <hr />

2. XML 태그는 대소문자를 구분함
  HTML에서는 태그 이름에 대소문자를 구분하지 않지만, XML은 구분하므로,
  대소문자가 다르면 다른 요소로 인식함

3. XML 태그의 여닫는 순서는 반드시 지켜져야 함

예제
<p><strong>이 구문은 오류를 발생합니다.</p></strong>

4. 속성값은 반드시 따옴표로 감싸야 함
  HTML에서는 속성값을 따옴표로 감싸지 않아도 큰 문제 없이 동작하지만
  XML에서는 오류 발생

예제
<student name=홍길동>   // 오류가 발생함.
<student name="이순신"> // 정상적으로 동작함.

6. XML에서는 띄어쓰기를 인식함

예제
코드 : <p>띄   어 쓰    기</p>
HTML : 띄어쓰기
XML  : 띄   어 쓰    기
```````````````

### XML 요소 문법

```````````````python
- XML 요소란 시작 태그부터 종료 태그까지의 모든 것을 가리킴
- XML 요소는 다음과 같은 문법을 사용하여 정의할 수 있음

문법
<요소이름 속성1="속성값" 속성2="속성값"...>내용</요소이름>

- 또한, 어떠한 내용도 가지지 않는 빈 요소(empty element)는 다음과 같이 정의할 수 있음

문법
<요소이름 속성1="속성값" 속성2="속성값".../>
```````````````

# 지도학습 알고리즘의 이해

## 머신러닝의 지도학습

```````````````python
- '학습할 데이터와 명시적인 정답(레이블)을 이용'해 데이터의 특성과
 분포를 학습하여 새로운 데이터에 대해 '미래결과를 예측'하는 방법

-- 분류(Classification)
  - k-최근접 이웃 알고리즘
  - 서포트 벡터 머신
  - 의사결정트리
  - 랜덤포레스트
  - 나이브 베이즈
-- 회귀(Regrssion)
  - 선형회귀
  - 로지스틱 회귀
```````````````

## K-최근접 이웃 알고리즘

```````````````python
k-Nearest Neighbors 알고리즘
- 이웃: 가까이 존재하는 데이터들을 의미
- 기존의 데이터 안에서 현재 데이터로부터 가까운 k개의 데이터를 찾아
 k개의 레이블 중 가장 많이 분류된 값으로 분류하는 알고리즘
- K : 최근접 점(일반적으로 홀수로 정함)    

최근접점 찾기
- 피타고라스 정리를 이용
```````````````

## 서포트 벡터 머신의 이해

```````````````python
Support Vector Machine
- 서로 다른 분류 값을 결정하는 경계선(결정 경계선 decision boundary)을
 결정하는 알고리즘
- 벡터의 의미 : 2차원 공간 상에 나타난 데이터 포인트
- 서포트 벡터 : 결정경계선과 가장 가까이 맞닿은 데이터 포인트
- 마진(margin) : 서포트 벡터와 결정 경계 사이의 거리
    
- SVM의 목표 : 마진을 최대로 하는 결정 경계를 찾는 것
 (마진이 클수록 알지 못하는 새로운 데이터에 대해 안정적으로 분류할 가능성이 높음)
- 마진을 최대로 하는 결정 경계는 학습 단계를 통해 발견

벡터 공간을 N차원의 데이터 벡터공간이라 한다면,
결정 경계 = N-1 차원.
```````````````

### 커널 트릭

```````````````python
저차원 벡터 공간 데이터(1차원)를 고차원 벡터 공간(2차원)으로 옮김
- 저차원일 때 정하기 힘들었던 결정 경계선을 정할 수 있게됨
```````````````

### 파라메터 최적화

```````````````python
# 마진을 얼마나 할 것인가?
- 비용(Cost): 마진조절 변수
 -- 비용이 작을수록 마진 너비가 넓어짐

- 마진을 크게 할 경우 학습 시 에러율이 높음

# 마진을 얼마나 할 것인가?
- 비용(Cost): 마진조절 변수
 -- 비용이 클수록 마진 너비가 좁아짐

- 마진을 작게 할 경우 학습 시 에러율은 낮음
- But, 학습되지 않은 데이터는 에러가 발생 될 가능성이 있음
```````````````

```````````````python
- 감마(Gamma)
 - 학습 데이터 포인트들이 결정 경계에 영향을 끼치는 범위를 
    조절해주는 변수
 - 감마 값이 크면 많은 데이터 포인트들이 가까이 있는 것으로 고려되어
  결정 경계가 작아지고 구부러짐
(감마 값이 작으면 데이터 포인트들이 멀리 분포되어 있는 것으로 고려돼
결정 경계가 완만해짐)
    
# 비용(Cost)와 감마(gamma) 수치를 조정하면 정확도를 높일 수 있음
```````````````

### 데이터 획득

```````````````python
- NBA 농구선수들의 게임 기록 데이터
(https://www.basketball-reference.com/leagues/NBA_2019_per_game.html)
- 스마트캠퍼스에서 다운 : basketball_state.csv
    
# NBA 농구선수들의 게임 기록 데이터 200개 항목

Player Pos  --선수이름, 포지션
3P  --한 경기 평균 3점 슛 성공 횟수
TRB --한 경기 리바운드 성공 횟수
BLK --한 경기 블로킹 성공 횟수
```````````````

### basketball_stat.csv 파일의 데이터를 DataFrame으로 불러오기

```````````````python
import pandas as pd

df = pd.read_csv(r'C:\Users\user\Documents\주피터노트북\data\basketball_stat.csv')
data_df = df[0:100]  # 슬라이싱
data_df.head()  # 5줄만 불러옴
```````````````

## 데이터 나누기

### 머신러닝 모델을 학습할 데이터 셋과 테스트할 데이터 셋으로 나누기

```````````````python
# 사이킷런 라이브러리의 train_test_split을 사용하여
# 학습 데이터와 테스트 데이터 분리

from sklearn.model_selection import train_test_split

# 학습 데이터 80%, 테스트데이터 20%로 분리
train, test = train_test_split(data_df,test_size=0.2)
train.shape

# Pos : 레이블(정답)
# 3P, TRB, BLK : 데이터
```````````````

### 학습 데이터셋에서 데이터와 레이블(정답)을 나누기

```````````````python
train_data_df= train[['3P','BLK','TRB']]
train_label_df= train[['Pos']]
train_data_df.head()
```````````````

```````````````python
# 인덱스와 열 이름이 있는데
# 머신러닝 알고리즘을 구현한 함수 안에 데이터를 넘겨줄 때,
# 인덱스와 열 이름을 제외한 코어 데이터만 필요함
train_data = train_data_df.values  # 값 부분만 가져오기(2차원 배열의 형태)
train_data
```````````````

```````````````python
train_label_df.head()
```````````````

```````````````python
train_label_df.values  # 80행 1열의 2차원 배열
```````````````

```````````````python
# 정답 데이터는 머신러닝에 넘길 때 코어 데이터를 1차원 데이터로 넘겨줘야 함
train_label = train_label_df.values.ravel() 
# 2차원에서 -> 1차원 배열 로 재구성
train_label
```````````````

```````````````python
test # 20개의 테스트 데이터
```````````````

```````````````python
test_data_df = test[['3P','BLK','TRB']]
test_label_df = test[['Pos']]

test_data = test_data_df.values
test_label = test_label_df.values.ravel()
```````````````

## kNN 모델 학습하기

### kNN 모델 학습하기

```````````````python
from sklearn.neighbors import KNeighborsClassifier
# KNN 라이브러리 추가

knn= KNeighborsClassifier(n_neighbors=3)
# KNN 분류기 객체 생성, 최근접 이웃의 수를 3으로 가정
knn.fit(train_data, train_label)
# knn 모델 학습
```````````````

### 테스트 데이터로 예측하기

```````````````python
pred_knn = knn.predict(test_data)
pred_knn
```````````````

## kNN 모델 예측 정확도 확인하기

```````````````python
from sklearn import metrics

ac_score = metrics.accuracy_score(test_label, pred_knn)
# 테스트 데이터셋의 레이블(정답)과 예측값 비교
print('accuracy:', ac_score)

# 수행 시 마다 다를 수 있음
```````````````
