### 여러 그래프 그리기(하나의 그래프 창에 여러 그래프 그리기)

```````````````python
# plt.plot(x1,y1[,ft1],x2,y2[,fmt2],x3,y3[,fmt3],...,xn,yn[,fmtn])
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5,5,0.1)
y1=2*x**2
y2=5*x+30
y3=20*np.cos(x)**2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
# plt.plot(x,y1,x,y2,x,y3)랑 동일한 결과

plt.show()
```````````````

### 새로운 그래프 창에 그래프 그리기

```````````````python
# plt.figure()    # 새로운 창을 생성

plt.plot(x,y1)    # 처음 그리기 함수 수행하면 그래프 창이 자동으로 생성됨

plt.figure()      # 새로운 그래프 창 생성
plt.plot(x,y2)    # 새롭게 생성된 그래프 창에 그래프를 그림

plt.figure()      # 새로운 그래프 창 생성
plt.plot(x,y3)    # 새롭게 생성된 그래프 창에 그래프를 그림

plt.show()
```````````````

```````````````python
%matplotlib qt  # 이 명령 수행 후 위의 그래프 다시 그려보면 3개의 그래프 창이 뜸

plt.plot(x,y1)   

plt.figure()     
plt.plot(x,y2)    

plt.figure()      
plt.plot(x,y3)    

plt.show()
```````````````

```````````````python
# 컬러 지정을 위한 약어

컬러 약어    컬러

b            파란색(blue)
g            녹색(green)
r            빨간색(red)
c            청녹색(cyan)
m            자홍색(magenta)
y            노란색(yellow)
k            검은색(black)
w            흰색(white)


# 선 스타일 지정을 위한 약어

선 스타일 약어    선 스타일

-                실선(solid line) ------------  # 일자로 쭉 이어진 선임!!!!!
--               파선(dashed line) - - - - - - -
:                점선(dotted line) `````````````` # 점선임
-.               파선 점선 혼합선(dash-dot line) -`-`-`-`-`-`-`-`


# 마커 지정을 위한 약어

마커 약어     마커

o             원모양●
^,v,<,>       삼각형 위쪽(▲), 아래쪽(▼), 왼쪽(◀), 오른쪽(▶)방향
s             사각형(square) ■
p             오각형(pentagon)
h,H           육각형(hexagon)1, 육각형2
*             별 모양(star)★
+             더하기(plus) +
x,X           x, 채워진 x
D,d           다이아몬드(diamond, ◆), 얇은 다이아몬드
```````````````

## 막대(bar) 그래프

```````````````python
# 여러 항복의 데이터를 서로 비교할 때 사용
#plt.bar(x, height[,width=width_f,color=colors, tick_label=tick_labels, align='center'(기본) 혹은 'edge'])

# x는 x축에 표시될 위치, heigh는 막대 그래프 데이터

import matplotlib.pyplot as plt
import numpy as np

member_IDs =['m_01','m_02','m_03','m_04']   # 회원 ID
before_ex =[30,40,42,35]        # 운동 시작 전
after_ex=[27,35,40,33]          # 운동 한 달 후

n_data = len(member_IDs)   # 회원이 네 명이므로 전체 데이터 수는 4
index = np.arange(n_data)  # numpy를 이용해 배열 생성 (0,1,2,3)
plt.bar(index, before_ex)  # x=index, height=before_ex로 지정
plt.show()       # x축의 index 위치에서 가운데 막대그래프 표시
```````````````

```````````````python
colors =['r','g','b','m']
plt.bar(index, before_ex, width =0.6, color=colors, tick_label=member_IDs)
# x축 표시 위치에 tick_label로 각각 막대그래프 이름 지정
plt.show()
```````````````
